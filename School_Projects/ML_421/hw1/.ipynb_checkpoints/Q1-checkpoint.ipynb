```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
```

```python
stock = pd.read_csv('StockBeta.csv')
```

```python
stock.head()
```

## 1.a

```python
# Plot IBM v.s. S_P500
plt.scatter(stock['S_P500'], stock['IBM'])
plt.xlabel('S_P500')
plt.ylabel('IBM')
```

```python
# Plot Apple v.s. S_P500
plt.scatter(stock['S_P500'], stock['Apple'])
plt.xlabel('S_P500')
plt.ylabel('Apple')
```

### The scatter plot of both S_P500 v.s. IBM and S_P500 v.s. Apple on the rate of return shows a positive correlation and linear relationship

## 1.b

```python
# Set stock return of IBM and Apple as target
target1 = stock['IBM']
target2 = stock['Apple']
```

```python
# Use S_P500 as the feature
X = sm.add_constant(stock['S_P500'])
```

```python
model1 = sm.OLS(target1, X).fit()
model2 = sm.OLS(target2, X).fit()
```

```python
# Regression for IBM
model1.summary()
```

### The slope for IBM regression is 0.7448

```python
# Regression for Apple
model2.summary()
```

### The slope for Apple regression is 1.2449, showing that with a unit increase in S_P500 index, Apple stock is expected to rise more compared to IBM's

## 1.c

```python
# Calculate SD for each stock sample
IBM_sd = np.std(target1)
Apple_sd = np.std(target2)
S_P_std = np.std(stock['S_P500'])
```

```python
print(f'sd for IBM: {IBM_sd} \nsd for Apple: {Apple_sd} \nsd for S_P500: {S_P_std}')
```

```python
# Calculate correlation matrix
stock.iloc[:, 1:4].corr()
```

```python
# Calculate beta hat for IBM and Apple using the formula given
IBM_beta_hat = 0.59 * 0.055 / 0.044
print(IBM_beta_hat)
Apple_beta_hat = 0.538 * 0.103 / 0.044
print(Apple_beta_hat)
```

### By comparison, the calculated beta_hats is approximately identical to that obtained from the regressions.

## 1.d

### According to the statistics calculated, the stock sample of Apple has a much higher sd of 0.103 compared to 0.055 of IBM. Also, the regression slope shows more variation than that of IBM, with a standard error of 0.193 compared to that of IBM regression of 0.099. Both statistics show that Apple stock bears much greater volatility for both its general trends and increasing patterns.
