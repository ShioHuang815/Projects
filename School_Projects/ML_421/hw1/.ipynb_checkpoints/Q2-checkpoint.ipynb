```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
```

```python
auto = pd.read_csv('auto.txt', sep="\s+")
```

```python
auto.head()
```

```python
# Check if there are nans in the features
auto['horsepower'].unique()
```

```python
auto['origin'].unique()
```

```python
# Set mpg to the target
target = auto['mpg']
# Fill nan in horsepower with means
auto['horsepower'] = auto['horsepower'].fillna(value = auto['horsepower'].mean())
```

```python
# Set horsepower and origin to the feature matrix
X = sm.add_constant(auto['horsepower'])
```

```python
# Regress on the target by X
model = sm.OLS(target, X).fit()
```

```python
# Print a summary of the model
model.summary()
```

```python
# Draw a scatterplot with a regression line, with different origins represented
sns.scatterplot(data = auto, x='horsepower', y='mpg', hue='origin', style='origin')
plt.plot(auto['horsepower'], model.predict(X))
```

## Q2.a: 
### The estimated equation for this regression is $$ y = 40.0058 - 0.1578(horsepower)$$

## Q2.b 
### For horsepower: The slope -0.1329 tells that for each unit of increase in horsepower, one could expect to see 0.1578 decrease in the mpg a car can drive

## Q2.c
### The regression slope has a standard error of 0.007, meaning it variates within a relatively small range, hence has limited uncertainty.

## Q2.d

```python
model.resid.std()
```

### A residual standard error of 4.98 tells that on average, the predicted values of the dependent variable deviate from the actual values by approximately 4.98 units. Considering this specific regression, this number shows a relatively good fit.

## Q2.e

### The regression slope has a p-value of approximately 0.000 according to the t-test, meaning that this regression shows certainty to reject the null hypothesis that horsepower has no effect on the mpg. Thus, this slope yield statistical significance.

## Q2.f

### R-square of 0.595 (0.594 for adjusted) shows about 59% of variation in mpg could be explain by this regression

## Q2.g

```python
model.predict((1, 98))
```

### The expected mpg for a car with horsepower of 98 is 24.53

## Q2.h

### The predicted mpg within a 95% prediction interval is according to the formula $$ \text{Prediction Interval} = \hat{y} \pm t_{\alpha/2} \cdot SE_{\text{pred}}$$

### Thus, in this case it is $$ 24.54 \pm 1.96 \times 4.98$$

## Q2.i

```python
# Calculate sample standard error
sd = np.std(target)
```

```python
# Find 99% confidence interval using t value of 2.626
CI = (24.54 - 2.626 * sd / len(target), 24.54 + 2.626 * sd / len(target))
```

## Q2.k

```python
sns.scatterplot(data = auto, x='horsepower', y='mpg', hue='origin', style='origin')
plt.plot(auto['horsepower'], model.predict(X))
```

### According to the graph, the regression shows heteroskedasticity. The variance of residuals is obviously greater with smaller horsepowers. Also, cars from difference origins indicates different trends. For example, those manufactured in location 3 shows greater variation compared to those from origin 1 and 2. This indicates that mere regression on horsepower may not be inclusive enough to account for prediction on mpg.
