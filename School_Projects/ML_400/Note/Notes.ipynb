## 1.1 AI-ML-DL

```python
# Create a figure and axis
fig, ax = plt.subplots()

# Draw the outer circle for AI (fill with blue color and transparent)
outer_circle = plt.Circle((0.5, 0.5), 0.45, color='blue', fill=True, alpha=0.3, label="AI")

# Draw the middle circle for ML (fill with green color and transparent)
middle_circle = plt.Circle((0.5, 0.5), 0.3, color='green', fill=True, alpha=0.3, label="ML")

# Draw the inner circle for DL (fill with red color and transparent)
inner_circle = plt.Circle((0.5, 0.5), 0.15, color='red', fill=True, alpha=0.3, label="DL")

# Add the circles to the plot
ax.add_artist(outer_circle)
ax.add_artist(middle_circle)
ax.add_artist(inner_circle)

# Add text to label the circles
ax.text(0.5, 0.875, 'AI', horizontalalignment='center', verticalalignment='center', fontsize=15, color='blue')  # Outer circle (AI)
ax.text(0.5, 0.7, 'ML', horizontalalignment='center', verticalalignment='center', fontsize=15, color='green')  # Middle circle (ML)
ax.text(0.5, 0.5, 'DL', horizontalalignment='center', verticalalignment='center', fontsize=15, color='red')   # Inner circle (DL)

# Set limits and aspect
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')  # Ensure the circle is not elliptical

# Hide axes
ax.axis('off')

# Show the plot
plt.show()
```
